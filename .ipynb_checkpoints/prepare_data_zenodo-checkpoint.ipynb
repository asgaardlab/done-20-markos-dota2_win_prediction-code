## Build and organize data to publish (Zenodo)

### Import libraries

```python
import os
import json
from time import gmtime
import statistics as st
import pandas as pd
import numpy as np
from shutil import copyfile
```

```python
# adjust cell width of the notebook 
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:75% !important; }</style>"))
```

```python
cwd = os.getcwd()
print(cwd)

# full dataset directory
dataset_dir = cwd + "\\data-collection\\matches-after-cleaning\\"
print(dataset_dir)

heroes_json_dir = cwd + "\\data-analysis\\"
print(heroes_json_dir)

# matches dict directory with historical stats
matches_dict_dir = cwd  + "\\data-collection\\matches_dict\\matches_dict_all\\"
print(matches_dict_dir)

# matches dict directory with historical stats (newdata: deaths/min and damage/min)
matches_dict_newdata_dir = cwd  + "\\data-collection\\matches_dict\\matches_dict_all-newdata\\"
print(matches_dict_newdata_dir)

zenodo_historical_stats_dir = cwd + "\\zenodo-dataset\\historical_statistics\\"
print(zenodo_historical_stats_dir)

zenodo_historical_stats_date_dir = cwd + "\\zenodo-dataset\\historical_statistics_date\\"
print(zenodo_historical_stats_date_dir)

hero_historical_stats_dir = cwd + "\\data-analysis\\hero_historical_stats\\"
print(hero_historical_stats_dir)

zenodo_historical_attributes_dir = cwd + "\\zenodo-dataset\\hero_historical_attributes\\"
print(zenodo_historical_attributes_dir)

```

```python
all_match_dict = os.listdir(matches_dict_dir)
print("number of match dict: " , len(all_match_dict))

all_match_dict_newdata = os.listdir(matches_dict_newdata_dir)
print("number of match dict newdata: " , len(all_match_dict_newdata))
```

```python
# open heroes json with constants
with open(heroes_json_dir + "heroes.json", 'r') as hero_file:
    heroes = json.load(hero_file) # 119 different heroes May 23rd
```

```python
list_hero_id = list()
for each_hero_id in heroes:
    list_hero_id.append(each_hero_id)

match_date_dict = {}
counter = 0

# all_match_dict = ['dict_1385942777.csv', 'dict_586003937.csv']
for match_dict in all_match_dict:
    print(counter)
    counter += 1
    match_id = match_dict.split('_')[1].split('.')[0]
    print(match_id)
    
    # open json match file
    with open(dataset_dir + "match-" + match_id + ".json", 'r') as match_file:
        match = json.load(match_file) # 119 different heroes May 23rd
        start_time = match['start_time']
        match_date = '{}-{}-{}'.format(*gmtime(start_time))
        match_date_dict[match_id] = match_date
    
    main_hist_df = pd.read_csv(matches_dict_dir + match_dict)
    main_hist_df = main_hist_df.drop(main_hist_df.columns[0], axis=1)

    hist_newdata_df = pd.read_csv(matches_dict_newdata_dir + match_dict)
    deaths_min_newdata = list(hist_newdata_df['deaths_min'])
    taken_damage_min_newdata = list(hist_newdata_df['health_damage_min'])
    
    # replace old values by new collected values
    main_hist_df.iloc[28:42, 4] = deaths_min_newdata
    
    # add new data (taken_damage_min_newdata)
    new_columns = ['id', 'win_rate', 'xp_min', 'gold_min', 'deaths_min', 'kills_min', 'assists_min', 'caused_damage_min',
                  'healing_min']
    main_hist_df.columns = new_columns
    
    # new column
    main_hist_df['taken_damage_min'] = [np.nan for x in range(28)] + taken_damage_min_newdata
    
    # change 'id' column indices
    id_list = list(main_hist_df['id'])
    new_id_indices = list()
    
    rad_heroes = id_list[:5]
    list_hero_names_rad = list()
    
    for hero_id in rad_heroes:
        for this_id in list_hero_id:
            if int(this_id) == int(hero_id):
                hero_dict = heroes[str(this_id)]
                hero_name = hero_dict['localized_name']
                list_hero_names_rad.append(hero_name)
                new_id_indices.append(hero_name.strip())
    
    new_id_indices.append("mean")
    new_id_indices.append("median")
                
    dire_heroes = id_list[7:12]
    list_hero_names_dire = list()
    
    for hero_id in dire_heroes:
        for this_id in list_hero_id:
            if int(this_id) == int(hero_id):
                hero_dict = heroes[str(this_id)]
                hero_name = hero_dict['localized_name']
                list_hero_names_dire.append(hero_name)
                new_id_indices.append(hero_name.strip())
                
    new_id_indices.append("mean")
    new_id_indices.append("median")
    
    rad_players = id_list[14:19]
    for each_ply in rad_players:
        new_id_indices.append(str(each_ply).strip())
    
    new_id_indices.append("mean")
    new_id_indices.append("median")

    dire_players = id_list[21:26]
    for each_ply in dire_players:
        new_id_indices.append(str(each_ply).strip())
    
    new_id_indices.append("mean")
    new_id_indices.append("median")

    rad_hero_player = id_list[28:33]
    ind = 0
    for hp in rad_hero_player:
        new_id_indices.append(str(int(hp)) + "-" + list_hero_names_rad[ind].strip())
        ind += 1

    new_id_indices.append("mean")
    new_id_indices.append("median")        

    dire_hero_player = id_list[35:40]
    ind = 0
    for hp in dire_hero_player:
        new_id_indices.append(str(int(hp)) + "-" + list_hero_names_dire[ind].strip())
        ind += 1

    new_id_indices.append("mean")
    new_id_indices.append("median")
    
    # update the 'id' column
    main_hist_df['id'] = new_id_indices
    
    main_hist_df[["id"]].style.hide_index().set_properties(**{'text-align': 'left'})

    
    # save new dataframe
    main_hist_df.to_csv(zenodo_historical_stats_dir + "match_" + match_id + ".csv", index=False)

```

```python
list_unique_date = list()
for key in match_date_dict:
    date = match_date_dict[key]
    if date not in list_unique_date:
        list_unique_date.append(date)
        
# create directories for each date
for date in list_unique_date:
    os.mkdir(zenodo_historical_stats_date_dir + date)
```

```python
for key in match_date_dict:
    match_id = key
    date = match_date_dict[key]
    copyfile(zenodo_historical_stats_dir + "match_" + match_id + ".csv", zenodo_historical_stats_date_dir + date + "\\" + "match_" + match_id + ".csv")
```

## Remove index ('unnamed') column in hero historical attribute files

```python
hero_historical_stats = os.listdir(hero_historical_stats_dir)
print("number of files: " , str(len(hero_historical_stats)))

for file in hero_historical_stats:                              
    hist_stats = pd.read_csv(hero_historical_stats_dir + file)
    hist_stats = hist_stats.drop(hist_stats.columns[0], axis=1)
    hist_stats.to_csv(zenodo_historical_attributes_dir + file, index=False)
    
```
